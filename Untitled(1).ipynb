## 对数值部分进行的可视化整理

```python
data=pd.DataFrame(pd.read_csv('../dataset/INvideos.csv'))
```

```python
data.info()
```

## 查看数据属性
#### 从数据描述中可以看出数据可以分为如下几类：
#### 1. 数值属性 包括 category_id, views, likes, dislikes, comment_count
#### 2. 布尔属性 包括 comments_disabled, ratings_disabled, video_error_or_removed
#### 3. 其余均是以字符串描述的属性

#### 本次处理选取数值属性进行处理

```python
for key in data.keys():
    print("{:>22}:{} missing values".format(key,data[key].isna().sum()))
```

## 对缺失值进行统计，可以看出只有description一项有缺失值

```python
fig,plots=plt.subplots(5,1,figsize=(10,20))
plots=plots.flatten()
labels=['category_id','views','likes','dislikes','comment_count']
for i in range(len(labels)):
    plots[i]=sns.distplot(data[labels[i]],ax=plots[i])
    plots[i].set_title("distribution of '{}'".format(labels[i]))

plt.tight_layout()
```

## 可以看出，除了category_id以外，其他的数值属性均存在分布不平衡的状态，因此这里我们进行一步取对数操作来使得分布更平衡，由于原数据集存在0值，因此这里统一加一进行处理

```python
data['views']=np.log(data['views']+1)
data['likes']=np.log(data['likes']+1)
data['dislikes']=np.log(data['dislikes']+1)
data['comment_count']=np.log(data['comment_count']+1)
```

```python
new_labels=['views','likes','dislikes','comment_count']
fig,plots=plt.subplots(4,1,figsize=(10,20))
for i in range(len(new_labels)):
    plots[i]=sns.distplot(data[new_labels[i]],ax=plots[i])
    plots[i].set_title("distribution of '{}'".format(new_labels[i]))

plt.tight_layout()
```

## 对标称属性进行计数，并统计计数结果

```python
plt.figure(figsize=(14,10))
g=sns.countplot('category_id',data=data,palette="Set1",order=data['category_id'].value_counts().index)
g.set_title('count of the video categories',fontsize=15)
g.set_ylabel("count",fontsize=12)
plt.show()
```

```python
wineOri = pd.DataFrame(pd.read_csv('../dataset/wine/winemag-data_first150k.csv'))
```

```python
wineOri.head()
```

```python
countries_list=pd.DataFrame(wineOri.country.value_counts(dropna=True))
```

```python
countries_list
```

```python
li=countries_list.index.values
```

```python
dic={}
i=0
for item in li:
    dic[item]=i
    i=i+1
```

```python
dic
```

```python
wineOri['country_id'] = wineOri['country'].map(dic)
```

```python
wineOri.head()
```

```python
province_list=pd.DataFrame(wineOri.province.value_counts(dropna=True))
province_list
```

```python
proli=province_list.index.values
prodic={}
i=0
for item in proli:
    prodic[item]=i
    i=i+1
prodic
```

```python
wineOri['province_id'] = wineOri['province'].map(prodic)
wineOri
```
